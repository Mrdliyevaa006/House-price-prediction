```python
import re
pattern=r'[0-9]+.[0-9]+|[0-9]{2,}'
def Area(value):
    splitted=value.split(' ')
    for i in splitted:
        if re.match(pattern,i):
            return i
df['Area(m2)']=df['Description'].apply(Area)
```

```python
import numpy as np
def Feature_Function(df: pd.DataFrame):
    target=df['Price']
    
    numerical_c = [c for c in df.select_dtypes(include=['int64', 'float64']) if df[c].nunique() != 2]
    
    new_dict = {}
    for c in numerical_c:
        
        feature_dict = {
            f'{c}>mean+std': df[c] > (df[c].mean() + df[c].std()),
            f'{c}<mean+std': df[c] < (df[c].mean() + df[c].std()),
            f'{c}_>mean': df[c] > df[c].mean(),
            f'{c}_<mean': df[c] < df[c].mean(),
            f'{c}_>q3': df[c] > df[c].quantile(0.75),
            f'{c}_<0.9': df[c] < df[c].quantile(0.9),
            f'{c}_log': np.log1p(df[c]),  
            f'{c}_sqrt': np.sqrt(df[c]),
        }
        for p in range(4, 10):
            feature_dict[f'{c}_power_{p}'] = np.power(df[c], p)
        
        for i in np.arange(0.1, 1.1, 0.1):
            feature_dict[f'{c}_power_{i}'] = np.power(df[c], i)

        real_corr = abs(df[c].corr(target))
        best_corr = real_corr
        best_feature = None

        for feature_name, feature_values in feature_dict.items():
            new_corr = abs(feature_values.corr(target))
            if new_corr > best_corr:
                print(f'New feature: {feature_name} | Original corr: {real_corr}, New corr: {new_corr}')
                best_corr = new_corr
                best_feature = feature_name

        if best_feature:
            new_dict[best_feature] = best_corr

    print('Best results:')
    return new_dict

new_features_dict = Feature_Function(df)
print(new_features_dict)
```

```python
from sklearn.ensemble import RandomForestRegressor
t='Price'
y = df[t]
x = df.drop(columns = [t])

reg = RandomForestRegressor(max_depth=6, random_state=0)

reg.fit( x, y )

x['pred'] = reg.predict(x)
x['pred'] = x['pred'].astype(int)
x['pred'] = x['pred'].round(-1)
x['real'] = y
x['error'] = x['pred'] - x['real']
x['abs-error'] = x['error'].abs()
x['percentage'] = x['abs-error'] / x['real']


print("CORR", x['pred'].corr(x['real']))
print( "MAE", x['abs-error'].mean() )
# MAE = mean absolute error
 
print("MAPE",x['percentage'].mean())
# MAPE = mean absolute percentage error

# MSE = mean square error
x['square'] = x['error'] * x['error']
print('MSE', x['square'].mean())

from sklearn.metrics import r2_score
print("R2", r2_score( x['real'], x['pred'] ))
#weighted_accuracy.append( r2_score( x['real'], x['pred'] ) * len(x) )
print()
```
